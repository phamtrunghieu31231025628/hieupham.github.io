```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display # hiển thị rich output
```

```python
N = 50
x = np.random.rand(N)          # 50 giá trị x trong [0,1]
y = np.random.rand(N)          # 50 giá trị y trong [0,1]
colors = np.random.rand(N)     # 50 giá trị màu trong [0,1]
area = np.pi * (15 * np.random.rand(N)) ** 2  # diện tích dựa trên bán kính ngẫu nhiên
```
