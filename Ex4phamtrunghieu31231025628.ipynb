```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sinh 256 điểm từ -pi đến pi
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

# In thông tin dữ liệu mẫu
print(f'+ X: |X|={len(X)}, x in [{np.min(X):.2f}, {np.max(X):.2f}], 10 samples: {X[0:10]}')
print(f'+ Cos(X): |C|={len(C)}, C in [{np.min(C):.2f}, {np.max(C):.2f}], 10 samples: {C[0:10]}')
print(f'+ Sin(X): |S|={len(S)}, S in [{np.min(S):.2f}, {np.max(S):.2f}], 10 samples: {S[0:10]}')
```

```python
# Vẽ đồ thị
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cos(x)")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sin(x)")

# Giới hạn trục X
plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# Giới hạn trục Y
plt.ylim(-1.1, 1.1)
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

# Thêm chú thích
plt.legend(loc="upper left")

# Hiển thị đồ thị
plt.show()
```
